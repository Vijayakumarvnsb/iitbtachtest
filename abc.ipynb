```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('abc.csv')
df.head()
```

```python
df.dtypes
```

```python
df_obj = df.select_dtypes(include=['object']).copy()
df_obj.head()
```

```python
df_obj[df_obj.isnull().any(axis=1)]
```

```python
df_obj['num_doors'].value_counts()
```

```python
obg_df = df_obj.fillna({"num_doors":'four'})
```

```python
obg_df.head()
```

```python
obg_df['num_cylinders'].value_counts()
```

```python
clean_up_find_reeplace = {"num_doors": {"four":4, "two":2},
                         "num_cylinders":{'four':4,'six':6,'five':5,'eight':8,'two':2, 'twelve':12,'three':3}}
```

```python
obj_df =  obg_df.replace(clean_up_find_reeplace)
```

```python
obj_df.head()
```

```python
obj_df.dtypes
```

```python
obj_df['body_style'].value_counts()
```

```python
obj_df['body_style'] = obj_df['body_style'].astype('category')
obj_df.dtypes
```

```python
obj_df['body_style_category'] =obj_df['body_style'].cat.codes
obj_df.head()
```

```python
obj_df['drive_wheels'].value_counts()
```

```python
pd.get_dummies(obj_df,columns=['drive_wheels'],prefix=['drive']).head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot = OneHotEncoder()
oone_hot_results = one_hot.fit_transform(obj_df[['drive_wheels']])
pd.DataFrame(oone_hot_results.toarray(), columns = one_hot.categories_).head()
```
